```python
sf3 = censusdata.download('sf3', 2000, censusdata.censusgeo([('state', '*')]),
                                   ['UA','H008001','H008002','H008003','H008004','H008005','H008006','H008007'])

sf3.columns = ['total_emp','rent','sale','rent_sold','season','migrant','other']
sf3
sf3.to_sql("emptyhouses", engine ,if_exists='replace',index=False)
```
